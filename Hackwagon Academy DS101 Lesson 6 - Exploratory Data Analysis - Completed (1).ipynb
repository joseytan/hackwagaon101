`Central Tendency`
>  - What is the mean, median resale price of a given (or each) town?

### Hint:
Use a dictionary to aggregate the data into the following format:
```python
{
    # town:       prices
    'ANG MO KIO': [100000, 200000, ...],
    'BUKIT BATOK': [200000, 70000, ...],

```
Then, loop over the dictionary and call the appropriate statistics function on the prices

Expected output:
```

  mean: $416695.36

BEDOK
  mean: $419418.30
  median: $385000.00
```

````python
import statistics

town_prices = {}

# Step 1: Accumulate towns and their prices into town_prices
for row in data:
    town = row["town"]
    resale_price = row["resale_price"]
    
    if town in town_prices:
        town_prices[town].append(resale_price)
    else:
        town_prices[town] = [resale_price]

# Step 2: Loop over town_prices and call statistics functions
for town, prices in town_prices.items():
    mean = statistics.mean(prices)
    median = statistics.median(prices)
    print(town)
    print("  mean: $" + str(round(mean, 2)))
    print("  median: $" + str(round(median, 2)))
    print()
````

`Distribution`
>  - What is the five number summary (min, q1, q2, q3, max) of the prices for resale flats in Seng Kang?

```
min = min of a set of numbers
q1 = 25th percentile of a set of numbers
q2 = 50th percentile of a set of numbers
q3 = 75th percentile of a set of numbers
max = max of a set of numbers
```

Hint: `np.quantile(list_of_numbers, 0.25)` gives `q1` 

Expected output
```
Resale Flat Prices
------------------
min: $196000.0
q1:  $403000.0
q2:  $450000.0
q3:  $515000.0
max: $720000.0

```

````python
import numpy as np

town = 'SENGKANG'

# Step 1: Accumulate prices (numbers) into a list
prices = []

for row in data:
    current_town = row["town"]
    resale_price = row["resale_price"]
    
    if current_town == town:
        prices.append(resale_price)
    
# Step 2: Call np.quantile on the list of numbers with the corresponding percentiles
minimum = np.quantile(prices, 0)
q1 = np.quantile(prices, 0.25)
q2 = np.quantile(prices, 0.5)
q3 = np.quantile(prices, 0.75)
maximum = np.quantile(prices, 1)

print("Resale Flat Prices")
print("------------------")
print("min: $" + str(minimum))
print("q1:  $" + str(q1))
print("q2:  $" + str(q2))
print("q3:  $" + str(q3))
print("max: $" + str(maximum))
````

````python
def five_summary_stats(town_to_check):
    # Step 1: Accumulate prices (numbers) into a list
    prices = []

    for row in data:
        current_town = row["town"]
        resale_price = row["resale_price"]

        if current_town == town_to_check:
            prices.append(resale_price)

    # Step 2: Call np.quantile on the list of numbers with the corresponding percentiles
    minimum = np.quantile(prices, 0)
    q1 = np.quantile(prices, 0.25)
    q2 = np.quantile(prices, 0.5)
    q3 = np.quantile(prices, 0.75)
    maximum = np.quantile(prices, 1)

    print("Resale Flat Prices")
    print("------------------")
    print("min: $" + str(minimum))
    print("q1:  $" + str(q1))
    print("q2:  $" + str(q2))
    print("q3:  $" + str(q3))
    print("max: $" + str(maximum))
    
five_summary_stats("SENGKANG")
five_summary_stats("ANG MO KIO")
````

### Q1. Each Town's Mean Price per Sqm

<p>
    <font color="grey">
        The price of a given flat is likely to be determined by its size, as we saw above. However, property prices are determined by a long list of other factors, such as proximity to amenities and public transport. Comparing the mean price per sqm of all flats sold in each town should give us a picture of how these towns are valued.
    </font>
</p>

> Determine mean `price per sqm` for all resale flats sold in each individual town.

Expected output:
```python
{'CENTRAL AREA': 7364.147617284561,
 'QUEENSTOWN': 6563.074741172909,
 'BUKIT MERAH': 6415.619384024125,
 ...}
```

````python
# here we aggregate to find the price_per_sqm for every resale flat for each town
# it will look like
# {
#     "ANG MO KIO": [5555.555556, 6022.7272...],
#     "BEDOK": [5533.333333, 5955.555556...],
#     ...
# }
town_prices_per_sqm = {}

for row in data:
    town = row["town"]
    price = row["resale_price"]
    area = row["floor_area_sqm"]
    
    # compute the price per sqm for this flat
    price_per_sqm = price / area

    if town in town_prices_per_sqm:
        town_prices_per_sqm[town].append(price_per_sqm)
    else:
        town_prices_per_sqm[town] = [price_per_sqm]

# after having the list of prices per sqm for each town,
# calculate mean price per sqm for each town
town_mean_ppsqm = {}

for town, prices in town_prices_per_sqm.items():
    town_mean_ppsqm[town] = sum(prices) / len(prices)
    
# we want to also sort the key-value pairs in town_mean_ppsqm
# by putting the town with the highest mean price per sqm first
# this is how to do it:
town_mean_ppsqm = {k: v for k, v in sorted(town_mean_ppsqm.items(), key=lambda x: x[1], reverse=True)}
# (there is no need to be too troubled by how it works, but feel free to reach out to me if you can't sleep without understanding it)

# display() is similar to print
# it is a special function that prints out the dictionary nicely
display(town_mean_ppsqm)


plt.figure(figsize=(12, 6))
plt.bar(town_mean_ppsqm.keys(), town_mean_ppsqm.values(), color='salmon')
plt.title('Mean Price per Sqm Per Town')
plt.ylabel('Resale Price ($)')
plt.xticks(rotation=90)
plt.show()
````

### Q2. Resale Price Estimate

<p>
    <font color="grey">
        Every couple / family intending to purchase a flat to settle down undoubtedly have constraints -- be it location, flat size, budget, etc. A programme that narrows down the search space for a suitable flat would be tremendously helpful for flat seekers. 
        
Many useful websites have search and filtering mechanisms to make users' lives easier.
    </font>

> Given a set of constraints, calculate the mean price of all flats possessing the respective attributes that satisfy them.

```python
town = 'Ang Mo Kio'
flat_type = '2 ROOM'
floor_area >= 30
```

Expected output:
```python
250175.5789...
```

````python
# these are the "constraints" or "requirements"
town = "ANG MO KIO"
flat_type = "2 ROOM"
min_floor_area = 30

prices = []

for row in data:
    # we only want prices that satisfy the following conditions
    cond1 = row["town"] == town
    cond2 = row["flat_type"] == flat_type
    cond3 = row["floor_area_sqm"] >= min_floor_area
    
    if cond1 and cond2 and cond3:
        prices.append(row["resale_price"])

print(sum(prices) / len(prices))
````
